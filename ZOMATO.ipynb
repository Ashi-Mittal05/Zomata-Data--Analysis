*Importing Libraries*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df= pd.read_csv('zzomato.csv')
```

```python
df.drop(['url','address','phone','dish_liked','reviews_list','menu_item'],axis=1,inplace=True)
```

```python
df.duplicated()
```

```python
df.duplicated().sum()
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].isnull().sum()
```

```python
df.dropna(inplace = True)
df.head()
```

```python
df.rename(columns = { 'listed_in(type)':'Rest_Type'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
df.drop('listed_in(city)',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['approx_cost(for two people)'].unique
```

```python
def handlecost(value):
 value = str(value)
 if ',' in value:
    value = value.replace(',', '')
    return float(value)
 else:
     return float(value)
    
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].apply(handlecost)
df['approx_cost(for two people)'].unique()
```

*Data Visualisation*

```python
sns.countplot(data=df,x='Rest_Type',color='red')
plt.xlabel("Type of Restaurant")
plt.title("Distribution of Restaurant Type")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.title("Online Vs. Offline Order Preference")
sns.countplot(data=df,x='online_order',palette='inferno')
```

```python
plt.figure(figsize=(15,10))
sns.countplot(data=df,x='location')
plt.title("Restaurant's Location")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data=df,x='online_order',hue='Rest_Type',palette='inferno')
plt.title("Distribution Of Online Orders")
```

```python
grouped_data = df.groupby ('Rest_Type')['votes']. sum()
result = pd. DataFrame({'votes': grouped_data})
plt. plot(result, c="maroon", marker="o")
plt.xlabel("Type of restaurant")
plt.xticks(rotation=90)
plt.title("Total Votes By Restaurant Type")
plt.show()
```

```python
plt.hist(data=df,x='rate', bins=5,color='purple')
plt.title("Ratings Distribution")
```

```python
sns.boxplot(x = 'book_table', y = 'rate', data = df,palette='rainbow')
plt.title("Impact of Table Booking on Restaurant Ratings")
```

```python
sns.boxplot(data=df, x="online_order", y="rate",palette='muted')
plt.title("Impact of Online Order on Restaurant Ratings")
```
